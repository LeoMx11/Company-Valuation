```python
def getHtml(url):

  page = requests.get(url, "html")

  soup = BeautifulSoup(page.text, 'html.parser')

  text = str(soup)

  return text
```

```python
check = 0

text1 = text

TickerList = []

while check == 0:

  text1 = text1[text1.find('<a href="/stocks/') + 1:]
  text1 = text1[text1.find('>')+1:]

  TickerList.append(text1[:text1.find('<')])

  if text1.find('<a href="/stocks/') == -1:
    check = 1

TickerList

len(TickerList)
```

```python
import yfinance as yf
import pandas as pd
```

```python
def mostRecentQtrDate(ticker):

  return str(yf.Ticker(ticker).get_financials(freq="quarterly").keys()[0])[:10]
```

```python
aaplData = yf.Ticker("BRK.B")

aaplData.financials
```

```python
bigDf = pd.DataFrame()

for stock in TickerList:

  concatThis = pd.DataFrame()

  try: concatThis[stock] = yf.Ticker(stock).get_financials(freq = "quarterly")[mostRecentQtrDate(stock)]
  except: pass

  try: bigDf = pd.concat([bigDf, concatThis], axis = 1)
  except: pass

bigDf
```

```python
bigDf = bigDf.fillna(0)

columnList = bigDf.columns

rowMeans = bigDf.mean(axis=1)

rowMeans
```

```python
len(columnList)

bigDf = bigDf.div(rowMeans, axis=0)

```

```python
import torch
```

```python
bigTensor = torch.from_numpy(bigDf.to_numpy().transpose()).type(torch.float)
```

```python
bigTensor
```

```python
PriceList = []

i=0

for ticker in list(columnList):

  try: PriceList.append((yf.Ticker(ticker).history(period="80d")["Open"].iloc[1]))
  except:
    print(ticker + " failed")
    bigTensor = torch.cat((bigTensor[:i], bigTensor[i+1:]))
    columnList = columnList.drop(ticker)

  i+=1

print(len(PriceList))
print(len(columnList))
```

```python
from sklearn.model_selection import train_test_split
```

```python
dataTrain, dataTest, priceTrain, priceTest = train_test_split(bigTensor, torch.tensor(PriceList), test_size=0.2, random_state=42)
```

```python
dataTrain = torch.nan_to_num(dataTrain, nan=0.0)
dataTest = torch.nan_to_num(dataTest, nan=0.0)
```

```python
priceTrain = priceTrain.type(torch.float)
priceTest = priceTest.type(torch.float)
```

```python
# import necessary stuff
import torch
from torch import nn

# make device agnostic code
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
# construct model
class StockPriceModel1(nn.Module):
  def __init__(self):
    super().__init__()

    self.layer1 = nn.Linear(in_features=84,
                            out_features=20)

    self.layer2 = nn.Linear(in_features=20,
                            out_features=5)

    self.layer3 = nn.Linear(in_features=5,
                            out_features=1)

    self.relu = nn.ReLU()

  def forward(self, x):
    return self.layer3(self.layer2(self.relu(self.layer1(x))))
```

```python
model_0 = StockPriceModel1().to(device)
model_0.state_dict()
```

```python
optimizer = torch.optim.SGD(params = model_0.parameters(),
                            lr=0.001)
loss_fn = torch.nn.functional.l1_loss
```

```python
def accuracy_fn(priceTrue, pricePred):

  total = 0

  for i in range(len(pricePred)):
    error = torch.abs(pricePred[i]-priceTrue[i]) / priceTrue[i]
    if error < 0.4:
      total += 1

  return total / len(pricePred)

```

```python
nn.functional.l1_loss(torch.tensor([3.0]), torch.tensor([5.0]))
```

```python
epochs = 1000

for epoch in range(epochs):

  model_0.train()

  #forward pass
  price_preds = model_0(dataTrain).squeeze()

  #calculate loss
  loss = loss_fn(price_preds, priceTrain)

  acc = accuracy_fn(priceTrue=priceTrain,
                   pricePred=price_preds)

  # optimizer zero grad
  optimizer.zero_grad()

  # perform backpropagation
  loss.backward()

  # optimizer step
  optimizer.step()

  # TESTING
  model_0.eval()
  with torch.inference_mode():

    price_test_preds = model_0(dataTest).squeeze()

    test_loss = loss_fn(price_test_preds,priceTest)
    test_acc = accuracy_fn(priceTest, price_test_preds)

  if epoch % 100 == 0:
    print(f"Epoch: {epoch} | Loss: {loss} | Acc: {acc}")
    print(f"Test Loss: {test_loss} | Test acc: {test_acc}")
```
